####  Removing non-digit and non-dot characters & converting the price column to numeric type

```python
allopathic_data['Retail_Price'] = allopathic_data['Retail_Price'].str.replace('[^\d.]', '', regex=True)
```

```python
allopathic_data['Retail_Price'] = pd.to_numeric(allopathic_data['Retail_Price'])
```

## Top manufecturers

```python
grouped_data = allopathic_data.groupby('Name_of_the_Manufecturer')
```

#### Based on the total retail price

```python
manufacturer_totals = grouped_data['Retail_Price'].sum()
```

```python
sorted_manufacturers = manufacturer_totals.sort_values(ascending=False)
top_manufacturers = sorted_manufacturers.head(10)
top_manufacturers
```

```python
market_share = manufacturer_totals / manufacturer_totals.sum()
```

```python
sorted_market_share = market_share.sort_values(ascending=False)
sorted_market_share.head(20).plot(kind='bar', title='Market Share by Manufacturer', figsize=(18,10))
```

#### Based on count of entries for each manufecturer

```python
manufacturer_counts = grouped_data.size()
sorted_manufacturers = manufacturer_counts.sort_values(ascending=False)
top_manufacturers = sorted_manufacturers.head(10)
top_manufacturers
```

#### Based on DAR i.e registered drugs

```python
manufacturer_dar = grouped_data['DAR'].count()
sorted_manufacturers = manufacturer_dar.sort_values(ascending=False)
top_manufacturers = sorted_manufacturers.head(10)
top_manufacturers
```

## Price Analysis

```python
# Distribution of Retail Prices
plt.figure(figsize=(10, 6))
plt.hist(allopathic_data['Retail_Price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Retail Prices')
plt.xlabel('Retail Price')
plt.ylabel('Frequency')
plt.show()
```

## Drug Usage Patterns:

```python
top_drugs = allopathic_data['Generic_Name'].value_counts().head(10)
plt.figure(figsize=(12, 8))
top_drugs.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Most Commonly Used Drugs')
plt.xlabel('Generic Name')
plt.ylabel('Frequency')
plt.show()
```
